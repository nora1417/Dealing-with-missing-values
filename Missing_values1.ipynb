What should we do about these missing values?

A- Dropna()


one common way is to drop the missing values.


1- drops a row if any of its values are missing.


losing almost all rows with a missing value/s.

```python
df.dropna(how='any').shape
```

```python
df.shape
```

2- Only drop a row if all of its values are missing.

It didn't drop any row.

```python
df.dropna(how='all').shape
```

3- dropoing rows in certain columns


```python
df.dropna(subset=['Lot.Number', 'Mill'], how='any').shape
```

Filling missing values.



B. Fillna()

1- Filling NANs with various


all the missing values were converted to "various".

```python
df['Lot.Number'].fillna(value='various', inplace=True)
df['Lot.Number'].value_counts()
```

```python
df.head()
```

2- Replacing NANs wiith 0.

```python
df_0 = df.fillna(value=0)
df_0
```

```python
df_0.isnull().sum()
```

3- Filling NANs with the previous value.(row wise)

```python
df_previous = df.fillna(method = 'pad')
df_previous
```

```python
df_previous.isnull().sum()
```

3.1- Filling NANs with the next value.(row wise)

```python
df_n = df.fillna(method = 'bfill')
df_n
```

```python
df_n.isnull().sum()
```

3.2 Filling NANs with the previous value(column wise).

```python
df_previous_col = df.fillna(method='pad', axis=1)
df_previous_col
```

```python
df_previous_col.isnull().sum()
```

3.3- Filling NANs with the next value(column wise).

```python
df_n_col = df.fillna(method = 'bfill', axis=1)
df_n_col
```

```python
df_n_col.isnull().sum()
```

4- Filling different NANs in different columns.


```python
df_col = df.fillna({'Lot.Number': 'abcd',
                    'Mill':'efgh'})
df_col
```

5- Filling NANs with the mean of a column.


```python
df_mean = df.fillna(value=df['Flavor'].mean())
df_mean
```

5.1- Filling NANs with the max of a column.


```python
df_max = df.fillna(value=df['Flavor'].max())
df_max
```

5.2- Filling NANs with the min of a column.

```python
df_min = df.fillna(value=df['Flavor'].min())
df_min
```

C. replace()

```python
df_replace = df.replace(to_replace=np.nan, value=123)
df_replace
```

D. interpolate()


fill the nan value with the mean of the other values.



```python
df_D = df['altitude_low_meters'] = df['altitude_low_meters'].interpolate(method='linear')
df_D
```

```python
df_D.isnull().sum()
```
